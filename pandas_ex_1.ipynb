```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
raw_fertility_rate_1950_2019 = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C", match="Коефіцієнт народжуваності в регіонах України")
```

```python
df = raw_fertility_rate_1950_2019[0]
```

Вивести перші рядки таблиці за допомогою методу head

```python
df.head()
```

Визначте кількість рядків та стовпців у датафреймі

```python
df.shape
```

Замініть у таблиці значення "—" на значення NaN

```python
df = df.replace("—", np.nan)
df
```

Визначте типи всіх стовпців за допомогою

```python
df.info()
```

Замініть типи нечислових колонок на числові

```python
df['2014'] = pd.to_numeric(df['2014'])
df['2019'] = pd.to_numeric(df['2019'])
df.info()
```

Порахуйте, яка частка пропусків міститься в кожній колонці

```python
df.isnull().sum()
```

Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df = df.drop(27, axis=0)
df
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців

```python
df["1950"].fillna(df["1950"].mean(), inplace=True)
df["1960"].fillna(df["1960"].mean(), inplace=True)
df["1970"].fillna(df["1970"].mean(), inplace=True)
df["2014"].fillna(df["2014"].mean(), inplace=True)
df["2019"].fillna(df["2019"].mean(), inplace=True)
df.isnull().sum()
```

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

##Порівняння народжуваності в 1950, 1960 та 1970 роках в регіонах України

```python
x = np.arange(df.shape[0])


fig, axs = plt.subplots()
axs.plot(df["Регіон"], df["1950"], label="Народжуваність за 1950 рік")
axs.plot(df["Регіон"], df["1960"], label="Народжуваність за 1960 рік")
axs.plot(df["Регіон"], df["1970"], label="Народжуваність за 1970 рік")

plt.setp(axs.get_xticklabels(), rotation=75, ha="right", rotation_mode="anchor")

plt.title("Порівняння народжуваності в 1950, 1960 та 1970 роках в регіонах України")
plt.ylabel("Кількість новонароджених малюків, тис.")
plt.legend(fontsize=10, ncol=4)
plt.grid()

plt.show()
```

##Діаграма загальної народжуваності по областям України

```python

fig, ax = plt.subplots(1, 1, figsize=(16, 9), dpi=80)
columns = df.columns[1:]
labs = columns.values.tolist()

x = df["Регіон"].tolist()
y0 = df["1950"].tolist()
y1 = df["1960"].tolist()
y2 = df["1970"].tolist()
y3 = df["1990"].tolist()
y4 = df["2000"].tolist()
y5 = df["2012"].tolist()
y6 = df["2014"].tolist()
y7 = df["2019"].tolist()

y = np.vstack([y0, y1, y2, y3, y4, y5, y6, y7])


ax.set_title("Діаграма загальної кількості малюків по областям України")
ax.legend(fontsize=10, ncol=4)

plt.ylabel("Кількість новонароджених малюків, тис.")
ax = plt.gca()
ax.stackplot(x, y,  labels=labs)

ax.legend(fontsize=10, ncol=4)
plt.xticks(x, fontsize=10, rotation=75, ha="right", rotation_mode="anchor")

plt.gca().spines["top"].set_alpha(0)
plt.gca().spines["bottom"].set_alpha(.3)
plt.gca().spines["right"].set_alpha(0)
plt.gca().spines["left"].set_alpha(.3)
plt.show()





```
