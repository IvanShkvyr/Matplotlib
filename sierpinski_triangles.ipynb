```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import random
```

```python
A = (3.5, 7)
B = (0, 0)
C = (7, 0)

apexs = (A, B, C)
```

Визначаємо координати точки D (за умовою точка D має лежати в межах трикутника ABC)

```python
def is_inside_triangle(A, B, C, D):
    AB = (B[0] - A[0], B[1] - A[1])
    AC = (C[0] - A[0], C[1] - A[1])
    AD = (D[0] - A[0], D[1] - A[1])
    cross1 = AB[0] * AD[1] - AB[1] * AD[0]
    cross2 = AC[0] * AD[1] - AC[1] * AD[0]

    if (cross1 < 0 and cross2 < 0) or (cross1 > 0 and cross2 > 0):
        return True
    else:
        return False

while True:
  D = (random.uniform(0, 7), random.uniform(0, 7))
  if is_inside_triangle(A, B, C, D):
    break
```

Створюємо набір координат для графіка "візерунок Серпінського"

```python
x = [A[0], B[0], C[0], D[0]] 
y = [A[1], B[1], C[1], D[1]]
for i in range(10000):
  P = random.choice(apexs)
  x.append((x[-1]+P[0])/2)
  y.append((y[-1]+P[1])/2)
```

```python
plt.scatter(x, y, s=1, alpha=0.5)
plt.title("Візерунок Серпінського", fontsize=20)
plt.gca().set_aspect('equal')
plt.gca().set_axis_off()
plt.rcParams["figure.figsize"] = [12, 8]
plt.show()
```
