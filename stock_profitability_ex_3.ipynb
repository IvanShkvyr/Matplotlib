```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy.stats import linregress
```

```python
data = pd.DataFrame({"stock_A": [25, -10, 10, 5, 35, 13],
                     "stock_B": [0, 15, -5, 5, 20, 25],
                     "stock_C": [10, 25, -15, -5, -5, 15]},
                     index=[1, 2, 3, 4, 5, 6])
```

```python

gridobj = sns.regplot(x="stock_A", y="stock_B", data=data)

slope, intercept, r_value, p_value, std_err = linregress(data["stock_A"], data["stock_B"])
equation = f"y = {slope:.2f}x + {intercept:.2f}"
correlation_coefficient = f"r = {r_value:.2f}"

plt.annotate(equation, xy=(0.05, 0.9), xycoords='axes fraction')
plt.annotate(correlation_coefficient, xy=(0.05, 0.85), xycoords='axes fraction')


plt.xlabel('прибутковість акцій A', fontsize='large')
plt.ylabel('прибутковість акцій B', fontsize='large')
plt.title("Залежність прибутковасті А акцій відносно В акцій", fontsize=20)
plt.rcParams["figure.figsize"] = [12, 8]
plt.show()
```

```python

gridobj = sns.regplot(x="stock_B", y="stock_C", data=data)

slope, intercept, r_value, p_value, std_err = linregress(data["stock_B"], data["stock_C"])
equation = f"y = {slope:.2f}x + {intercept:.2f}"
correlation_coefficient = f"r = {r_value:.2f}"

plt.annotate(equation, xy=(0.05, 0.9), xycoords='axes fraction')
plt.annotate(correlation_coefficient, xy=(0.05, 0.85), xycoords='axes fraction')

plt.xlabel('прибутковість акцій B', fontsize='large')
plt.ylabel('прибутковість акцій C', fontsize='large')
plt.title("Залежність прибутковасті B акцій відносно C акцій", fontsize=20)
plt.rcParams["figure.figsize"] = [12, 8]
plt.show()
```
